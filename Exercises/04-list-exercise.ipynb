````python
import random

simulated_dice_rolls_list = []

for _ in range (10):
    simulated_dice_rolls = random.randint(1, 6)
    simulated_dice_rolls_list.append(simulated_dice_rolls)

count_six = simulated_dice_rolls_list.count(6)

print (simulated_dice_rolls_list)
print (count_six)
````

````python
import random
import matplotlib.pyplot as plt

random.seed(1)


num_rolls = [10, 100, 1000, 10000, 100000, 1000000]
count_sixes = []
probabilities = []

for rolls in num_rolls:
    dice_rolls = [random.randint(1, 6) for _ in range(rolls)]
    count_six = dice_rolls.count(6)
    count_sixes.append(count_six)
    probability = count_six / rolls
    probabilities.append(probability)

for i, rolls in enumerate(num_rolls):
    print(f"In {rolls} rolls, the number 6 appeared {count_sixes[i]} times, with a probability of {probabilities[i]:.5f}")

plt.plot(probabilities, '-*')
plt.title("Probability of six for different number of rolls")
plt.xticks([0,1,2,3,4,5], num_rolls)
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")

````

## 6. Monte Carlo simulation (**)

&nbsp; a) Simulate 5000 uniformly distributed points between -1 and 1, $X\sim U(-1,1)$. Now calculate the Euclidean distances between each point and the origin (0,0), $d_i = \sqrt{x_i^2 + y_i^2}$. For $d_i < 1$ plot the point with one color, and for $d_i > 1$ plot the point in another color, $i = \{1,\ldots, 5000\}$.  (**)

&nbsp; b) Calculate the fraction between number of inner points and number of outer points. What do you this will converge to when number of points reaches infinity? (**)


<details>

````python
import random
import math
import matplotlib.pyplot as plt

num_points = 50000
inside_circle = []
outside_circle = []

for _ in range(num_points):
    x, y = random.uniform(-1, 1), random.uniform(-1, 1)
    distance = math.sqrt(x**2 + y**2)
    if distance <= 1:
        inside_circle.append((x, y))
    else:
        outside_circle.append((x, y))

x_inside, y_inside = zip(*inside_circle)
x_outside, y_outside = zip(*outside_circle)

plt.scatter(x_inside, y_inside, color='green')
plt.scatter(x_outside, y_outside, color='red')
plt.show()

num_inside = len(inside_circle)
num_outside = len(outside_circle)
fraction = num_inside / (num_inside + num_outside)
print(f"Fraction of points inside the circle: {fraction:.4f}")
````

## 7. A cute rabbit among two ferocious snakes (**)

There are three identical doors, behind two of them, there are ferocious snakes and behind one of them is a cute little rabbit. Your aim is to pick a door and get the cute rabbit. You start picking one door randomly and suddenly another door opens and snake slithers away *pssss*. Now you have a choice to either stay with your door or change to the remaining door. 

<img src="../Assets/snakes_rabbits.png" width = 200>

&nbsp; a) What is your choice and why? (**)

&nbsp; b) Randomly assign which door has the rabbit, and randomly pick an initial door. Now simulate 10, 100, 1000, 10000, 100000, 1000000 times for staying and for switching doors. Calculate the proportion of finding the rabbits for each set of simulations. Plot the results (**)


<details>

<summary>Hint</summary>

a) 

Try to reason using probability.


</details>

<details>

<summary>Answer</summary>


a)

b) 

<img src="../Assets/mh_simulations.png" width = 300>



</details>

````python
import random
import matplotlib.pyplot as plt

try_times = [10, 100, 1000, 10000, 100000, 1000000]
stay_proportions = []
switch_proportions = []

for n in try_times:
    stay_wins = 0
    switch_wins = 0
    for _ in range(n):
        doors = {"D1": "Snake", "D2": "Snake", "D3": "Snake"}
        door_with_rabbit = random.choice(list(doors.keys()))
        doors[door_with_rabbit] = "Rabbit"

        chosen_door = random.choice(list(doors.keys()))

        available_doors = [door for door in doors.keys() if door != chosen_door and doors[door] == "Snake"]
        revealed_door = random.choice(available_doors)

        remaining_door = [door for door in doors.keys() if door != chosen_door and door != revealed_door][0]

        if doors[chosen_door] == "Rabbit":
            stay_wins += 1
        if doors[remaining_door] == "Rabbit":
            switch_wins += 1

    stay_proportions.append(stay_wins / n)
    switch_proportions.append(switch_wins / n)


# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(try_times, stay_proportions, marker='o', label="Stay")
plt.plot(try_times, switch_proportions, marker='o', label="Switch")
plt.xscale('log')
plt.xlabel('Number of Simulations')
plt.ylabel('Proportion of Wins')
plt.title('Proportion of Wins for Staying vs. Switching')
plt.legend()
plt.grid(True)
plt.show()

````
